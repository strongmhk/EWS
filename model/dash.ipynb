```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd
import numpy as np
import pandas as pd
import os

os.chdir('/Users/hiksang/Desktop/Proj/DF/data')
df = px.data.iris()
```

```python
df = pd.read_csv('KVALL.WD.csv')

feature_col = df.iloc[:,4:].columns
feature_list = feature_col.to_list()
feature_bank = feature_list + ['bank.code']
feature_all = feature_bank + ['date']

def prep_data(df, feature_col):
    df = df[feature_col]
    df = df.replace([np.inf, -np.inf], np.nan)  # "inf" 값을 NaN으로 대체
    df = df.dropna()
    return df

df= prep_data(df, feature_list)
print(df)
```

```python
def box_plot(df, feature_list):
    fig = px.box(df, y=feature_list)
#     fig.show()
    
    return fig

box_plot(df, feature_list)
```

```python
# fig = px.scatter(df, x="sepal_length", y="sepal_width", 
#                   color="species")

app = Dash(__name__)
# app layout: html과 dcc 모듈을 이용
app.layout = html.Div(children=[
    # Dash HTML Components module로 HTML 작성 
    html.H1(children='입력받은 데이터로 Dashboard'),
    html.Div(children='''
        선택한 column으로 ploting
    '''),
    # dash.core.component(dcc)의 그래프컴포넌트로 plotly 그래프 렌더링
    dcc.Graph(
        id='graph1',
        figure=box_plot(df, ['KV001', 'KV002'])
    )
])

if __name__ == '__main__':
#     app.run_server(debug=True)
    app.run_server(debug=True, host='0.0.0.0', port=8100)
```

```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd

app = Dash(__name__)

# assume you have a "long-form" data frame
# see https://plotly.com/python/px-arguments/ for more options
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
    "Amount": [4, 1, 2, 2, 4, 5],
    "City": ["SF", "SF", "SF", "Montreal", "Montreal", "Montreal"]
})

fig = px.bar(df, x="Fruit", y="Amount", color="City", barmode="group")

app.layout = html.Div(children=[
    html.H1(children='Hello Dash'),

    html.Div(children='''
        Dash: A web application framework for your data.
    '''),

    dcc.Graph(
        id='example-graph',
        figure=fig
    )
])

if __name__ == '__main__':
    app.run(debug=True)
```
